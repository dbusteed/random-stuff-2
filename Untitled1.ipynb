```python
import pandas as pd
import shap
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
```

```python
df = pd.read_csv("../data/insurance_dummies.csv")
```

```python
X = df.drop(columns=['charges'])
y = df[['charges']]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2112)
```

```python
xgb = XGBRegressor()
xgb.fit(X_train, y_train)
```

```python
explainer = shap.Explainer(xgb)
```

```python
shap_values = explainer(X_test)
```

```python
vals = pd.DataFrame(shap_values.values, index=X_test.index, columns=X_test.columns)
vals.insert(loc=0, column='base', value=shap_values.base_values)
```
